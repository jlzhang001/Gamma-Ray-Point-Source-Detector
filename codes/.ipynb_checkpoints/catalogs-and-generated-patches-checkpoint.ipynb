#### Read fits catalogs and return the corresponding atlas

The following function reads the fits file map and create a list with the content of the SKYMAP structure, it also computes the sum of the content, which we interpret as the number of photons

```python
def get_sky_maps(map_IEM, map_AGN, map_PSR):
    in_HDU_IEM_BKGRD = map_IEM
    in_HDU_IEM_BKGRD = fits.open(in_HDU_IEM_BKGRD) 
    in_HDU_IEM_BKGRD[1].header['COORDSYS'] = 'G'
    nSide = in_HDU_IEM_BKGRD[1].header['NSIDE']
    order = in_HDU_IEM_BKGRD[1].header['ORDERING'] #e.g nested
    nPix = hp.nside2npix(nSide)
    EBIN = len(in_HDU_IEM_BKGRD[1].data[0])
    
    in_HDU_AGN = map_AGN
    in_HDU_AGN = fits.open(in_HDU_AGN) 
    in_HDU_AGN[1].header['COORDSYS'] = 'G'
    
    in_HDU_PSR = map_PSR
    in_HDU_PSR = fits.open(in_HDU_PSR) 
    in_HDU_PSR[1].header['COORDSYS'] = 'G'
        
    hdu_atlas_IEM = []
    hdu_atlas_AGN = []
    hdu_atlas_PSR = []
    
    for i in range(EBIN):
        in_hdu_IEM_chart= np.array(list(map(lambda x: x[i], list(in_HDU_IEM_BKGRD[1].data)))) ##extract sky map of i-th energy bin
        hdu_atlas_IEM.append(in_hdu_IEM_chart)
        in_hdu_AGN_chart= np.array(list(map(lambda x: x[i], list(in_HDU_AGN[1].data)))) ##extract sky map of i-th energy bin
        hdu_atlas_AGN.append(in_hdu_AGN_chart)
        in_hdu_PSR_chart= np.array(list(map(lambda x: x[i], list(in_HDU_PSR[1].data)))) ##extract sky map of i-th energy bin
        hdu_atlas_PSR.append(in_hdu_PSR_chart)
        
    return (hdu_atlas_IEM, hdu_atlas_AGN, hdu_atlas_PSR)
```

#### During training we must compute the solid angle correction considering xsize_patch_generation = 64, NSIDE=256 and add a factor 1/4 in order to recover the proper conversion

This is because the original resolution to produce the patches was 128x128, therefore in principle, the solid angle correction should takes this patch size and NSIDE=256. This produces a factor 1/8. However, during the new generation of patches we can consider xsize_patch_generation = 64 with solid angle correction included

```python
xsize_patch_generation = 64
NSIDE = 256

Npix = 12 * NSIDE * NSIDE
pix_sr = 4.0 * np.pi / Npix
```

```python
lb_range = [-5, 5]
#(-4.9609375, 4.9609375)
        
proj = hp.projector.CartesianProj(lonra=lb_range, latra=lb_range, xsize=xsize_patch_generation)

I,J = np.meshgrid(np.arange(xsize_patch_generation), np.arange(xsize_patch_generation))
x,y = proj.ij2xy(I,J)

dl = np.radians((lb_range[1]-lb_range[0])/(xsize_patch_generation-1))
solid_area_ratio = dl*dl*np.cos(np.radians(y))
```

#### Now, let us check that the corrected patch contains a similar amount of photons as the 64x64 corrected

```python
x_a = patch_np_128
global_correction_factor = solid_area_ratio/pix_sr/4

x_a_t = x_a[0,:,:,:] + x_a[1,:,:,:] + x_a[2,:,:,:]
                
for nbin in range(5):
    x_a_t[:,:,nbin] = x_a_t[:,:,nbin]*global_correction_factor
```

```python
def visualize_patch_5bins(grid):
    fig, ax = plt.subplots(5,3, figsize=(30, 50))
    
    for bin in range(5):  
               
        ax[bin][0].imshow(grid[0,:,:,bin])
        ax[bin][1].imshow(grid[1,:,:,bin])
        ax[bin][2].imshow(grid[2,:,:,bin])
```

#### Read fits catalogs and return the corresponding atlas

The following function reads the fits file map and create a list with the content of the SKYMAP structure, it also computes the sum of the content, which we interpret as the number of photons

```python
def get_sky_maps(map_IEM, map_AGN, map_PSR):
    in_HDU_IEM_BKGRD = map_IEM
    in_HDU_IEM_BKGRD = fits.open(in_HDU_IEM_BKGRD) 
    in_HDU_IEM_BKGRD[1].header['COORDSYS'] = 'G'
    nSide = in_HDU_IEM_BKGRD[1].header['NSIDE']
    order = in_HDU_IEM_BKGRD[1].header['ORDERING'] #e.g nested
    nPix = hp.nside2npix(nSide)
    EBIN = len(in_HDU_IEM_BKGRD[1].data[0])
    
    in_HDU_AGN = map_AGN
    in_HDU_AGN = fits.open(in_HDU_AGN) 
    in_HDU_AGN[1].header['COORDSYS'] = 'G'
    
    in_HDU_PSR = map_PSR
    in_HDU_PSR = fits.open(in_HDU_PSR) 
    in_HDU_PSR[1].header['COORDSYS'] = 'G'
        
    hdu_atlas_IEM = []
    hdu_atlas_AGN = []
    hdu_atlas_PSR = []
    
    for i in range(EBIN):
        in_hdu_IEM_chart= np.array(list(map(lambda x: x[i], list(in_HDU_IEM_BKGRD[1].data)))) ##extract sky map of i-th energy bin
        hdu_atlas_IEM.append(in_hdu_IEM_chart)
        in_hdu_AGN_chart= np.array(list(map(lambda x: x[i], list(in_HDU_AGN[1].data)))) ##extract sky map of i-th energy bin
        hdu_atlas_AGN.append(in_hdu_AGN_chart)
        in_hdu_PSR_chart= np.array(list(map(lambda x: x[i], list(in_HDU_PSR[1].data)))) ##extract sky map of i-th energy bin
        hdu_atlas_PSR.append(in_hdu_PSR_chart)
        
    return (hdu_atlas_IEM, hdu_atlas_AGN, hdu_atlas_PSR)
```
